# Approach 3: RGB, multispectral, and height from Sept 2 and Jul 2, with composite features

Composite features: the difference between the same feature on Sept 7 and Jul 2

```python
# import and clean Jul2 data
dataJul = pd.read_csv('module2_data/CleanData_InputML_7_2_19.txt',sep='\t')
dataJul.drop(labels = ['Accession','Genotype','AltGenotype','EarLeafArea'],axis = 1,inplace=True)
dataJul.head()
```

```python
dataSept = data2
dataSept.head()
```

```python
# add in multispectral data 
multispecMeans = pd.read_csv('module2_data/multispec_means_obs2019.csv')
multispecMeans.head()
```

```python
# clean out non-experimental measurements
excludeFill = []
IDs = multispecMeans.id
for name in IDs:
    if 'fill' in name:
        excludeFill.append(name)

multispecMeansExp = multispecMeans[~multispecMeans['id'].isin(excludeFill)]
multispecMeansExp.head()
```

```python
from collections import defaultdict
```

```python
# clean plot id's to match RGB
replacements = defaultdict(str)
for name in multispecMeansExp.id:
    splitStr = name.split('-')
    replacements[name] = splitStr[0]
    
multispecMeansExp.replace(replacements,inplace=True)
multispecMeansExp.head()
```

```python
# select dates 
dates = []
for col in multispecMeansExp.columns.values:
    if col != 'id' and '7_2_19' not in col:
        if '9_2_19' not in col:
            dates.append(col)
multispecMeansExp.drop(dates,axis=1,inplace=True)
multispecMeansExp.head()
```

```python
multispecMeansExp.drop(['COUNT_7_2_19','AREA_7_2_19','COUNT_9_2_19','AREA_9_2_19'],axis=1,inplace=True)
multispecMeansExp.head()
```

```python
# separate into jul and sept dfs 
sept = []
jul = []
for col in multispecMeansExp.columns.values:
    if col != 'id' and '7_2_19' in col:
        jul.append(col)
    elif col != 'id' and '9_2_19' in col:
        sept.append(col)
        
septMultispec = multispecMeansExp.drop(jul,axis=1)
julMultispec = multispecMeansExp.drop(sept,axis=1)
```

```python
julMultispec.head()
```

```python
septMultispec.head()
```

```python
# combine RGB and multispec for each date 
septMultispec.rename(columns = {'id':'PlotID'},inplace=True) # it was all because of this.......
julMultispec.rename(columns = {'id':'PlotID'},inplace=True)
```

```python
septMultispec.head()
```

```python
dataSept.head()
```

```python
multiID = septMultispec.PlotID
RGBid = dataSept.PlotID 
```

```python
membership = multiID.isin(RGBid)
membership.value_counts()
```

According to the above, there should be 356 columns lost when I merge over the PlotID column.

```python
septAll = pd.merge(septMultispec,dataSept,on='PlotID',how='inner')
septAll.head()
```

```python
# do it for july!
julMultispec.head()
```

```python
dataJul.head()
```

```python
julAll = pd.merge(julMultispec,dataJul,on='PlotID',how='inner')
julAll.head()
```

```python
# pair up column names for subtraction 
septCols = pd.Series(septAll.columns.values) 
julCols = pd.Series(julAll.columns.values)
subCols = pd.concat([septCols,julCols],axis=1)
subCols.drop([0,6],axis=0,inplace=True)
subCols
```

```python
# create full feature table 
AllFeatures = pd.merge(julAll,septAll,on='PlotID',how='inner')
AllFeatures.head()
```

```python
# perform subtraction and add columns onto the end of comined df 
for i in subCols.itertuples(): 
    sept = i[1]
    jul = i[2]
    featureName = 'diff_'+sept+'_'+jul
    newCol = septAll[sept] - julAll[jul]
    newCol.name = featureName
    AllFeatures = pd.concat([AllFeatures, newCol], axis=1)
    
```

```python
AllFeatures.head()
```

```python
# manually verify one of the differences 
blueJulZero = AllFeatures.iloc[0,1]
blueSeptZero = AllFeatures.iloc[0,28]
blueDiff = AllFeatures.iloc[0,55]
blueDiff == blueSeptZero - blueJulZero
```

```python
# check that the two TLA columns match and then get rid of one 
x = AllFeatures.TotalLeafAreaPlot_x
y = AllFeatures.TotalLeafAreaPlot_y
x.equals(y)
```

```python
AllFeatures.drop('TotalLeafAreaPlot_x',axis=1,inplace=True)
```

```python
AllFeatures.rename(columns = {'TotalLeafAreaPlot_y':'TotalLeafAreaPlot'},inplace=True)
```

```python
# rearrange so TLA is the second column 
featCols = AllFeatures.columns.values.tolist()
print(len(featCols))
groundIndex = featCols.index('TotalLeafAreaPlot')
groundTruth = featCols[groundIndex]
featCols.pop(groundIndex)
print(len(featCols))
featCols.insert(1, groundTruth)
featCols
```

```python
AllFeatures = AllFeatures[featCols]
AllFeatures.head()
```

```python
# rename columns to be compatible with pipeline
AllFeatures.rename(columns = {'PlotID':'ID','TotalLeafAreaPlot':'Class'},inplace=True)
AllFeatures.head()
```

```python
# write out as csv
AllFeatures.to_csv('module2_data/allButHeightFeatures.csv',index=False)
```
